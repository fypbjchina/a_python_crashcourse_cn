文件处理的基本概念
- 文件类型：
  - 文本文件：包含可读的字符（如 .txt， .csv， .py, .ipynb, .bat）。
  - 二进制文件：包含非文本数据（如图像、音频文件等）。

- 文件模式：
    - 'r'：只读模式（默认模式），文件必须存在。
    - 'w'：写入模式，创建新文件或覆盖现有文件。
    - 'a'：追加模式，向文件末尾添加内容。
    - 'b'：二进制模式（与其他模式结合使用，如 'rb' 或 'wb'）。
    - 'x'：独占写入模式，文件必须不存在。
      
文件处理的基本步骤
1. 打开文件：使用 open() 函数打开文件。
2. 操作文件：读取或写入文件内容。
3. 关闭文件：使用 close() 方法关闭文件，释放资源。
