```python
my_dict_copy1["Age"] = 27
print (my_dict)
print(my_dict_copy1)
```

```python
my_dict = dict({'Name': '123',
 'Age': 28,
 'ID': '123456678',
 'Char': {'FamilyRoll': ['代车司机', '课外辅导', '无条件买单', '猛男']},
 'Wealth': ['1000 BitCom', '1 Mil. Euro', '1000 Ans GOld']})
my_dict
```

```python
# deepcopy from module copy
import copy
```

```python
my_dict_dc1= copy.deepcopy(my_dict)
print(my_dict)
print(my_dict_dc1)
print(my_dict_dc1 == my_dict)
```

```python
my_dict_dc1["Age"] = 27
print(my_dict)
print(my_dict_dc1)
print(my_dict_dc1 == my_dict)
```
